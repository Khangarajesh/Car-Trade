```python
df = pd.read_csv("/content/car1.csv")
df.head()
```

**Data Cleaning**

*Dirty Data*:
*   Price column containing unnecessary ₹ and , `validity`
*   price column in string `validity`
*   kms_driven containing KMs and , `validity`
*   kms_driven is in string  `validity`
*   actual_pr containing nan values  `completeness`
*   actual_pr containing ₹ and , `validity`
*   unnecessary column Unnamed  `validity`
*   emi column containing nan value `completeness`
*   emi column contaning unnecessary string 'EMI starts at', ₹ and , `validity`
*   emi column has data type string `validity`




*Messy Data*:
*   Name column containing manufacturing year, car name and version


```python
#Cleaning name column

#Extract year from name
df['mfg_year'] = df['name'].str[:4]
```

```python
#Remove blank spaces
df['name'] = df['name'].str[4:].str.strip()
```

```python
#spliting name column
df['new'] = df['name'].str.split()
```

```python
df['new'].apply(lambda x: x[:3]).value_counts()
```

After observing above series I observe that first two elements of the list contains most of the model name and remanining is vrsion name.In few cases model name are till 3rd element

3rd element containing model names are

Celerio , Baleno, Wagon, Ertiga, i10, Eeco, Estilo,i20

```python
#Extracting model name 
df['new'] = df['new'].apply(lambda x : x[:3])
df['car_model'] = df['new'].apply(lambda x: ' '.join(x) if x[-1] in (['Celerio', 'Baleno', 'Wagon', 'Ertiga', 'i10', 'Eeco', 'Estilo', 'i20']) else ' '.join(x[:2]))
```

```python
#Extracting version
df['version'] = df['name'].str.split().apply(lambda x: ' '.join(x[2:]) if x[2] not in (['Celerio', 'Baleno', 'Wagon', 'Ertiga', 'i10', 'Eeco', 'Estilo', 'i20']) else ' '.join(x[3:]))
```

```python
df.drop(['Unnamed: 0', 'name', 'new'], axis = 1, inplace = True)
```

```python
df
```

```python
#cleaning price column
df['price'] = df['price'].str.replace('₹', '').str.replace(',', '').astype('int')
```

```python
#cleaning kms_driven column
df['kms_driven'] = df['kms_driven'].str.replace(',', '').str.replace(' KMs', '').astype('int')
```

```python
#emi column
df['emi'] = df['emi'].str.replace('EMI starts at ₹', '').str.replace(',', '')
```

By observation I found that most of the emi are for 66 month with 9.99% rate of intrest.(rate of interest for used car was given 9.99% on CarTrade website)

so we will replace nan values in emi column with value calulated based on 66 month and 9.99% rate of interes

```python
df['emi'] = df['emi'].astype('float')
df['emi'] = df['emi'].fillna(df.apply(lambda x : round(x['price']/66 * 1.0999,2), axis = 1))
```

```python
#actual price
df['actual_pr'] = df['actual_pr'].str.replace('₹', '').str.replace('Lakh', '').str.strip().astype('float')
#convert into a lakhs
df['actual_pr'] = df['actual_pr']*100000
#fill nan value with price column
df['actual_pr'] = df['actual_pr'].fillna(df.apply(lambda x: x['price'], axis = 1))
```
